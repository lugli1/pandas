Observação: Na nova atualização da biblioteca Pandas, a coluna de valores proporcionais, quando no dataframe, será nomeada automaticamente como proportions. Com essa mudança, é necessário alterar também a coluna selecionada para a criação da variável df_percentual tipo:

```python
df['Tipo'].value_counts(normalize=True).to_frame().sort_values('proportion')
```

```python
df_percentual_tipo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('proportion')
```

```python
df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color ='green', edgecolor='black',
                        xlabel = 'Tipos', ylabel = 'Percentual');
```

```python
#Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevadas.
aluguel_bairros = df.groupby('Bairro')[['Valor']].mean().sort_values('Valor', ascending=False)

aluguel_bairros.plot(kind="barh", figsize=(14, 10), color ='black',
                        xlabel = 'Bairros', ylabel = 'Média');
```
