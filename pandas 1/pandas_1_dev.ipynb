```python
import pandas as pd
```

```python
url= 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
```

```python
dados = pd.read_csv(url,sep=';')
```

```python
dados.fillna(0,inplace=True)
```

```python
dados['valor_por_mes'] = dados['Valor'] + dados['Condominio']
```

```python
dados['valor_por_ano'] = dados['valor_por_mes']*12 + dados['IPTU']
```

```python
dados.head(20)
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
url2 = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
```

```python
df = pd.read_csv(url2, sep=',')
```

```python
df.head()
```

```python
df = df.fillna(0)
```

```python
df['Pontos_extras'] = df['Notas']*0.4
```

```python
df.head()
```

```python
df['Notas_finais'] = df['Notas']+df['Pontos_extras']
```

```python
df.head()
```

```python
df['Aprovado_final'] = df['Notas_finais']>=6
```

```python
df.head(20)
```

```python
selecao = df.query('Aprovado == 0 & Aprovado_final == 1')
selecao
```
